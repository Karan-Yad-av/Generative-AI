2. How does broadcasting work in NumPy ?
   
   Broadcasting in NumPy is a powerful mechanism that allows NumPy to perform element-wise operations on arrays of different shapes without explicitly copying data.
*  Broadcasting lets NumPy automatically expand the smaller array so that both arrays have compatible shapes during an operation.
```
(2*3)+(1*3)
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

print(a + b)
Output : (2*3)
[[11 22 33]
 [14 25 36]]
```
```
(3*1)+(1*3)
a = np.array([[1],
              [2],
              [3]])
b = np.array([10, 20, 30])
print(a + b)
Output : (3*3)
[[11 21 31]
 [12 22 32]
 [13 23 33]]
```

17. Explain the difference between apply() and map() in Pandas .
* map() in Pandas
  * Works only on a Series
  * Applies a function element-wise
  * Used for simple transformations
  * Can accept : Function , Dictionary , Series
  ```
    import pandas as pd
    s = pd.Series([1, 2, 3])
    s.map(lambda x: x * 2)
  ```
  Cannot be used directly on DataFrames

* apply() in Pandas
  * Works on Series and DataFrames
  * Applies a function along rows or columns
  * Used for more complex operations
  * Can operate on:
      * Each column (axis=0)
      * Each row (axis=1)
  ```
  df.apply(sum)
  df.apply(lambda row: row["Math"] + row["Science"], axis=1)
  ```

21.  Why is NumPy’s array slicing faster than Python’s list slicing.

     NumPy slicing is faster because it creates a view of the data using contiguous memory and C-level operations, whereas Python list slicing creates a new list and copies elements one by one.
  * View vs Copy
      * NumPy slicing returns a view .
      * Python list slicing creates a new list .
  * Contiguous Memory Storage
      * NumPy arrays are stored in continuous blocks of memory
      * Python lists store references to objects scattered in memory
  * No Element-by-Element Iteration
      * NumPy adjusts metadata (start, stop, stride) only
      * Python lists loop through elements and copy them
  * Lower Memory Overhead
      * NumPy slicing uses almost no extra memory
      * Python list slicing duplicates references → more memory use
    ```
        import numpy as np
        a = np.arange(1000000)
        b = a[100:200]   

        lst = list(range(1000000))
        sub = lst[100:200]
    ```

````python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "Math": [85, 90, 78, 92],
    "Science": [88, 85, 80, 95],
    "English": [78, 82, 75, 88]}
df = pd.DataFrame(data)
corr_matrix = df.corr()
print(corr_matrix)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
````
