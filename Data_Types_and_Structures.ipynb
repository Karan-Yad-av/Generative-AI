13. How do sets handle duplicate values in Python ?

* In Python, a set automatically removes duplicate values.
* When you add elements to a set, only unique values are stored, and any repeated value is ignored.
    * Sets are unordered collections of unique elements.
    * If you try to add a duplicate element, Python simply does not store it again.
    * No error is raised when duplicates are added.

    ```
    s = set()
    s.add(10)
    s.add(10)
    s.add(20)
    print(s)  # >> output = (10,20)
    ```
    Sets handle duplicate values by automatically removing them, ensuring that only unique elements are stored.



14. How does the “in” keyword work differently for lists and dictionaries ?

    The in keyword is used to check for membership, but it behaves differently for lists and dictionaries.
    # "in" with Lists
      * Checks whether a value exists in the list.
      * It searches through the list elements one by one.
          ```
          list00 = [10, 20, 30]
          print(20 in list00)   # True
          ```

    # "in" with dictionaries
      * Checks whether a key exists, not the value.
      * It does not search values unless explicitly asked.
          ```
          dict11 = {"a": 1, "b": 2}
          print("a" in dict11) # true(for item keys)
          print(1 in dict11.values())  # true(for item values)
          ```

19.  Why are dictionaries considered unordered, and how does that affect data retrieval ?

     **Dictionaries** are considered unordered because they are implemented using hash tables.
     In a hash table, keys are placed in memory based on their hash values, not on the order in which they are added. Because of this, the internal storage order does not matter.
      * Data retrieval is not affected by order.

      * Values are accessed using keys, not by position.

      * Lookup time is very fast (O(1) on average).
      
     Dictionaries are considered unordered because they store data using hash tables, where elements are placed based on hash values rather than order. This does not affect data retrieval because values are accessed using keys, not positions.
