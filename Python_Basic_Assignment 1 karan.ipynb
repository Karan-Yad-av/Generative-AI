6.  Why are lists mutable, but tuples are immutable ?
    -   Lists are mutable because Python allocates flexible memory for them, allowing modification.
    -   Tuples are immutable because they use fixed memory, making their values unchangeable after creation.

  Memory Design :
  
      - List : Lists use a dynamic memory structure, meaning Python allocates extra space so elements can be changed, added, or removed.
      - Tuples : Tuples use a fixed memory structure, so once they are created, the values cannot change.


7. What is the difference between “==” and “is” operators in Python ?
  
    - "==" : (Equality Operator)

         - Checks whether values of two objects are the same.

         -  It compares contents.
        - Example :=>
        a = [1, 2, 3]   b = [1, 2, 3]
          print(a == b)  # output = True

    - "is" : (Identity Operator)

        - Checks whether two variables refer to the same memory location (same object).
        - It compares identity, not value.
        - Example :=> a = [1, 2, 3]  b = [1, 2, 3] print(a is b) # output = False

9. What is type casting in Python ?

      - Type casting in Python means converting a value from one data type to another.
      - It allows you to change the type of a variable—for example, converting a string to an integer or a float to a string.
      - Python provides built-in functions for type casting.

      **Implicit Type Casting (Automatic)** : Python automatically converts one data type to another without you asking, usually during arithmetic operation.

      **Explicit Type Casting (Manual)** : You manually convert one type to another using functions like :  int() , float() , str()

      Example :

        1.   Converting integer to float
              a = 10        # integer
              b = 3.5       # float
              c = a + b     # convert automatically
              print(c)      # 13.5 type is float

        2.  Converting string to integer
                  x = "25"
                  y = int(x)
                  print(y)     # 25

        3.  Converting integer to float
                  a = 10
                  b = float(a)
                  print(b)     # 10.0

13.  What is the difference between for and while loops ?

     **Loop** : A loop in Python is a control flow statement that allows a block of code to be executed repeatedly based on a condition or over a sequence of items.

        - **for loops** : Used to iterate over a sequence (like a list, tuple, string, or range) for a predetermined number of times.

           - A for loop is ideal for processing every item in a collection or running code a specific number of times using the range() function.
           
           Example :
           
                  for i in range(5):
                  print(i)

        - **while loops** : Used to repeat a block of code as long as a given condition remains true; the number of iterations is often unknown beforehand.  

           -  A while loop is used when the continuation of the loop depends on a condition that is evaluated dynamically at runtime.
          
          Example :
                
                i = 0
                while i < 5:
                     print(i)
                     i += 1


14.  Describe a scenario where a while loop is more suitable than a for loop.

      A while loop is more suitable when you do not know in advance how many times the loop should run, and the repetition depends on a condition that may change during the program.
       
      Example :

                password = "admin123"
                user_input = ""
                while user_input != password:
                user_input = input("Enter password: ")

                print("Access granted!")

     A while loop is more suitable when the number of repetitions is not known beforehand, such as repeatedly asking a user for valid input until they provide the correct response.
