6. Explain the concept of generators in Python and how they are defined.
   * A generator in Python is a special type of function that produces values one at a time instead of returning all values at once.
   * Generators are used to iterate over large data sets efficiently because they generate values on demand.
   * Generators are defined like normal functions, but they use the yield keyword instead of return.

      * yield returns a value and pauses the function’s state

      * Execution resumes from where it stopped when the next value is requested.
      ```
        def count(n):
        for i in range(n):
          yield i
        for num in count(5):
          print(num)
      ```



7. What are the advantages of using generators over regular functions ?
  * Generators provide several important benefits :
      * Memory Efficiency : Very useful for large data sets or infinite sequences.
      * Faster Startup : Generator functions start execution immediately and pause after yield.
      * Automatic State Management.


        ```
            def squares_list(n):
                return [i*i for i in range(n)] # Regular Function

            def squares_gen(n):
                for i in range(n):
                   yield i*i         # Generator
        ```


  Generators are more memory-efficient and faster than regular functions because they generate values on demand using yield.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 * **map()** : Applies a function to each element of an iterable and returns the transformed elements.
    * map fuction transform the elements.
    * return the multiple values as a output.
      ```
      numb = [1, 2, 3, 4]
      res = list(map(lambda x: x * 2, numb))
      print(res)
      ```
 * **filter()** : Selects elements from an iterable that satisfy a condition.
     * filter fuction select the elements.
    * return the fewer values as a output.
      ```
      numb = [1, 2, 3, 4]
      res = list(filter(lambda x: x % 2 == 0, numb))
      print(res)
      ```
  * **reduce()** : Repeatedly applies a function to reduce an iterable to a single value.
      * reduce fuction combine the elements.
    * return the single values as a output.
      ```
      numb = [1, 2, 3, 4]
      res = reduce(lambda x, y: x + y, numb)
      print(res)
      ```

  Conculsion : map() transforms elements, filter() selects elements based on a condition, and reduce() reduces all elements into a single value.

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
num = int(input("Enter a number b/w 1 and 200 : "))
if 1 <= num <= 200:
    if is_prime(num):
        print(num, "is Prime .")
    else:
        print(num, "is NOT Prime .")
else:
    print("Number out of range .")
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibbo:
    def gen(self, n):
        a, b, i = 0, 1, 1
        res = []
        while i <= n:
            res.append(a)
            a, b = b, a + b
            i+=1

        return res

n = int(input("Enter number of terms: "))
fib = Fibbo()
print(fib.gen(n))
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_2(n):
    for i in range(n + 1):
        yield 2 ** i
expo = int(input("Enter the exponent element : "))
for value in power_2(expo):
    print(value,end=" ")
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read(out):
  for line in out:
    yield line.strip()
out = ["                             Generators are commonly",
       "                 used in data processing            ",
       " file handling        "]
for l in read(out):
  print(l,end=" ")
````

````python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 3), (4, 1), (2, 2), (5, 0)]
s_data = sorted(data, key=lambda x: x[1], reverse=True)
print("Second element of the tuple in decreasing order : ",s_data)
s_data = sorted(data, key=lambda x: x[1], reverse=False)
print("Second element of the tuple in increasing order : ",s_data)
````
